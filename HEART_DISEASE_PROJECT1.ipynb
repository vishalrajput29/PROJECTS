# DATA DESCIPTION
This heart disease dataset contains 1,025 entries with 14 columns, each representing a distinct attribute or feature relevant to the diagnosis of heart disease. Below is a detailed description of each column:

age: Age of the patient in years (integer).

sex: Gender of the patient (integer: 1 = male, 0 = female).

cp (chest pain type): Type of chest pain experienced (integer: 0-3).

trestbps (resting blood pressure): Resting blood pressure in mm Hg (integer).

chol (serum cholesterol): Serum cholesterol level in mg/dl (integer).

fbs (fasting blood sugar): Fasting blood sugar > 120 mg/dl (integer: 1 = true, 0 = false).

restecg (resting electrocardiographic results): Resting ECG results (integer: 0-2).

0: Normal
1: Having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)

2: Showing probable or definite left ventricular hypertrophy by Estes' criteria
thalach (maximum heart rate achieved): Maximum heart rate achieved during exercise (integer).

exang (exercise-induced angina): Exercise-induced angina (integer: 1 = yes, 0 = no).
oldpeak: ST depression induced by exercise relative to rest (float).

slope: Slope of the peak exercise ST segment (integer: 0-2).

ca: Number of major vessels (0-3) colored by fluoroscopy (integer).

thal: Thalassemia status (integer: 1-3).


target: Diagnosis of heart disease (integer: 1 = presence of heart disease, 0 = absence of heart disease).

# **UNIVARIANT ANALYSIS**

```python

plt.figure(figsize = (10, 6))
sns.histplot(df['age'].dropna(), kde = True, color = 'skyblue')
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize = (10, 6))
sns.histplot(df['target'].dropna(), kde = True, color = 'skyblue')
plt.title('Distribution of target')
plt.xlabel('target')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (20, 15))
plotnumber = 1


for col in num_cols:
    if plotnumber <= 15:
        ax = plt.subplot(3, 5, plotnumber)
        sns.distplot(df[col])
        plt.xlabel(col)
    plotnumber += 1

plt.tight_layout()
plt.show()
```

# **BIVARIANT ANALYSIS**

```python
corr_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

HEART DISEASE WITH GENDER

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='sex', hue='target', palette='Set1')
plt.title('Distribution of Heart Disease by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Disease', loc='upper right', labels=['No Heart Disease', 'Heart Disease'])
plt.grid(True)
plt.show()
```
